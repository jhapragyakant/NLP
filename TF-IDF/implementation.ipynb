```python
import pandas as pd
df= pd.read_csv('D:\\NLP Temp\\NLP\\NLP\\BOW\\spamhamdata.csv',sep='\t', names= ['label', 'message'])
```

```python
## Data cleaning and preprocessing

import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer= WordNetLemmatizer()
```

```python
## This task is common and generic for all the usecases

corpus=[]
for i in range(len(df)):
    review= re.sub('[^a-zA-Z]', ' ', df['message'][i]) # deleting all the special characters and keeping only the alphabets
    review= review.lower() # lowering the text
    review= review.split() # createda a list of review
    review= [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')] #if the word is not present in stopwords, we gonna lemmatize it and store it agian in our review variable
    review= ' '.join(review)
    corpus.append(review)
```

### TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer as TFIDF
tfifd= TFIDF(max_features=100, binary=True, ngram_range=(2,2))
```

```python
X= tfifd.fit_transform(corpus).toarray()
```
